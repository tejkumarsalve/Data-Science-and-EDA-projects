Train and test the model

```python
# Traing the model by using train_test_spli
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into 9:1 ratio
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y)
```

```python
print(Y.mean(),Y_train.mean(),Y_test.mean())
```

```python
print(X.mean(),X_train.mean(),X_test.mean())
```

```python
print(X_train)
```

Scaling the data for better performance

```python
# Scalling the model by uisng StandardScalar
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

Logistic Regression model
It is usefull for binary distribution

```python
# Loading the Logistic Regression Model Nad fitting the data to the model
from sklearn.linear_model import LogisticRegression
Classifier=LogisticRegression()
Classifier.fit(X_train,Y_train)
```

Evaluation of model

```python
from sklearn.metrics import accuracy_score
```

```python
# Accuracy of training data
prediction_on_training_data=Classifier.predict(X_train)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
print("Accuracy on training data",accuracy_on_training_data)
```

```python
# Accuracy on test data
prediction_on_test_data=Classifier.predict(X_test)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
print("Accuracy on test data",accuracy_on_test_data)
```

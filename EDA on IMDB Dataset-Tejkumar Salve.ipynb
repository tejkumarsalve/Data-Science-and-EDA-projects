```python
# Converting the string data into list format for manipulation
def convert_to_str(string):
  return eval(string)
```

**4] Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?bold text**

```python
#for the list of producers and directors with highest average roi 
imdb_data_for_movie=imdb_data1.copy()
imdb_data_for_movie.head()
```

```python
imdb_data_for_movie['ROI']=100*imdb_data_for_movie['revenue'] / imdb_data_for_movie['budget']
imdb_data_for_movie.head()
```

```python
imdb_data_for_movie=imdb_data_for_movie[['original_title','cast','crew','ROI']]
imdb_data_for_movie.head()
```

```python
# function to obtain list of producers and directors of movie which has given highest Roi
def get_producer_and_director(crew_list):
  try:
# Empty producer and director list
    producer_list=[]
    director_list=[]
    for k in crew_list:
      if k['job']=='Producer':
         producer_list.append(k['name'])
      if k['job']=='Director':
         director_list.append(k['name'])

    return producer_list, director_list

  except:
    return [], []  
```

```python
# dataframe with new columns of producer and director using lambda 
imdb_data_for_movie[['Producer','Director']]=imdb_data_for_movie.apply(lambda z: get_producer_and_director(z['crew']),axis=1,result_type='expand')
```

```python
imdb_data_for_movie.head()
```

```python
# Creating a empty dataframe with columns
Table=pd.DataFrame(columns=['Movie_title','Producer','Director','ROI'])
```

```python
# appending the values to the empty 'Table' dataframe
for ind,rows in imdb_data_for_movie.iterrows():
 
  Table=Table.append({'Movie_title':imdb_data_for_movie.loc[ind,'original_title'],'Producer':imdb_data_for_movie.loc[ind,'Producer'],'Director':imdb_data_for_movie.loc[ind,'Director'],'ROI':imdb_data_for_movie.loc[ind,'ROI']},ignore_index=True)
 
```

```python
# Obtaing highest ROI row using sort_values()
Table=Table.sort_values('ROI',ascending=False)
Table.head()
```

Samuel L Jackson and Robert De Niro both acted in same number of movies which greater than any other actor

```python
# Profit and Movie by actors who have done most number of movies
sam_profit=[]
rob_profit=[]
sam_movie=[]
rob_movie=[]
for index,rows in imdb_data_for_actor.iterrows():
  for elem3 in imdb_data_for_actor.loc[index,'cast']:
    if type(elem3)==dict:
      actor_name=elem3['name']
      if 'Samuel L. Jackson' in actor_name:
        sam_profit.append(imdb_data_for_actor.loc[index,'profit'])  
        sam_movie.append(imdb_data_for_actor.loc[index,'original_title'])


      if 'Robert De Niro' in actor_name:
        rob_profit.append(imdb_data_for_actor.loc[index,'profit'])  
        rob_movie.append(imdb_data_for_actor.loc[index,'original_title'])
```

```python
#genres of actors with most number of movies
gen_sam=[]
for i in range(len(sam_movie)):
  for j in imdb_data_for_actor.loc[i,'genres']:
    gen_sam.append(j['name'])


gen_rob=[]
for i in range(len(rob_movie)):
  for j in imdb_data_for_actor.loc[i,'genres']:
    gen_rob.append(j['name'])
```

```python
# creating array from list
genre_rob=np.array(gen_rob)
genre_sam=np.array(gen_sam)
```

```python
# creating a dataframe with single column 
sam_movie1=pd.DataFrame(sam_movie,columns=['Movies'])
rob_movie1=pd.DataFrame(rob_movie,columns=['Movies'])
sam_profit1=pd.DataFrame(sam_profit,columns=['Profit'])
rob_profit1=pd.DataFrame(rob_profit,columns=['Profit'])
sam_genre1=pd.DataFrame(genre_sam,columns=['Genres'])
rob_genre1=pd.DataFrame(genre_rob,columns=['Genres'])
```

```python
# Concatinating the dataframes for obtaining the whole dataframe
Data_of_Samuel=pd.concat([sam_movie1,sam_genre1,sam_profit1],axis=1)
Data_of_Robert=pd.concat([rob_movie1,rob_genre1,rob_profit1],axis=1)
```

```python
#Movie of samuel L Jackson 
Data_of_Samuel.sort_values(by='Profit',ascending=False).head()
```
